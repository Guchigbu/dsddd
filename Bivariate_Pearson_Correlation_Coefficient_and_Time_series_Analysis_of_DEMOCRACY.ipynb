```python
#TIME and percentage of PERSON METAPHORS in Nigerian Political discourse using Spearman Rank

import pandas as pd
from scipy import stats

# Create the dataset
data = {
    'Year': [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020],
    'Value': [0.00, 0.00, 0.00, 0.84, 5.05, 6.73, 17.66, 6.73, 41.21, 16.82, 16.82, 8.41, 13.46]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Spearman correlation
spearman_corr, p_value = stats.spearmanr(df['Year'], df['Value'])

# Print results
print(f"Spearman's rank correlation coefficient: {spearman_corr:.3f}")
print(f"p-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("\nThe correlation is statistically significant (p < 0.05)")
else:
    print("\nThe correlation is not statistically significant (p ≥ 0.05)")
```

```python
import pandas as pd
from scipy import stats

# Create the dataset with BUILDING metaphor values
data = {
    'Year': [1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020],
    'Building_Metaphor': [0.00, 0.00, 0.84, 0.84, 13.46, 17.66, 61.40, 31.96, 44.58, 34.48, 38.69, 8.41, 19.34]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Spearman correlation
spearman_corr, p_value = stats.spearmanr(df['Year'], df['Building_Metaphor'])

# Print results with more detailed interpretation
print("=== Spearman Rank Correlation Analysis ===")
print(f"Correlation coefficient (ρ): {spearman_corr:.3f}")
print(f"p-value: {p_value:.5f}")

# Detailed interpretation
print("\nInterpretation:")
if p_value < 0.05:
    print(f"The correlation is statistically significant (p = {p_value:.5f})")
    if spearman_corr > 0.7:
        strength = "strong positive"
    elif spearman_corr > 0.3:
        strength = "moderate positive"
    else:
        strength = "weak positive"
    print(f"There is a {strength} monotonic relationship between year and BUILDING metaphor frequency (ρ = {spearman_corr:.3f}).")
else:
    print(f"The correlation is not statistically significant (p = {p_value:.5f})")

# Additional visualization suggestion
print("\nNote: Consider visualizing this relationship with:")
print("- A scatterplot with a trendline")
print("- A bar chart showing values by year")
```
