```python
#TIME SERIES ANALYSIS and JOURNEY METAPHOR (NATIONHOOD) in Nigerian Political discourse (Discarded)


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox

# Load and prepare data
data = [
    1960,5,1961,0,1962,1,1963,2,1964,0,1965,0,1966,4,1967,0,1968,0,1969,0,
    1970,27,1971,0,1972,3,1973,1,1974,8,1975,2,1976,3,1977,1,1978,9,1979,3,
    1980,4,1981,4,1982,2,1983,0,1984,32,1985,11,1986,20,1987,3,1988,5,1989,9,
    1990,0,1991,5,1992,1,1993,28,1994,17,1995,7,1996,16,1997,17,1998,2,1999,24,
    2000,15,2001,0,2002,17,2003,16,2004,4,2005,17,2006,3,2007,4,2008,1,2009,1,
    2010,15,2011,24,2012,1,2013,15,2014,32,2015,14,2016,5,2017,3,2018,6,2019,10,
    2020,10,2021,12,2022,5,2023,10,2024,18
]

# Reshape data into DataFrame
df = pd.DataFrame(np.array(data).reshape(-1, 2), columns=['Year', 'Count'])
df['Year'] = df['Year'].astype(int)
df['Count'] = df['Count'].astype(int)
df.set_index('Year', inplace=True)
df.index = pd.to_datetime(df.index, format='%Y')

# Plot original time series
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Count'], marker='o')
plt.title('Journey Metaphor Usage Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()

# Check stationarity with Augmented Dickey-Fuller test
adf_result = adfuller(df['Count'])
print(f'ADF Statistic: {adf_result[0]:.3f}')
print(f'p-value: {adf_result[1]:.3f}')
print('Critical Values:')
for key, value in adf_result[4].items():
    print(f'   {key}: {value:.3f}')

# Differencing if non-stationary
if adf_result[1] > 0.05:
    df_diff = df.diff().dropna()
    print("\nTaking first difference to achieve stationarity")
else:
    df_diff = df.copy()

# Plot ACF/PACF
fig, ax = plt.subplots(2, 2, figsize=(12,8))
plot_acf(df['Count'], lags=20, ax=ax[0,0])
plot_pacf(df['Count'], lags=20, ax=ax[0,1])
plot_acf(df_diff['Count'], lags=20, ax=ax[1,0])
plot_pacf(df_diff['Count'], lags=20, ax=ax[1,1])
plt.tight_layout()
plt.show()

# Ljung-Box test for autocorrelation
lb_test = acorr_ljungbox(df['Count'], lags=[5,10], return_df=True)
print("\nLjung-Box Test Results:")
print(lb_test)

# Fit ARIMA model (example order - adjust based on your ACF/PACF)
model = ARIMA(df['Count'], order=(1,1,1))
results = model.fit()
print("\nARIMA Model Summary:")
print(results.summary())

# Forecast next 5 years
forecast = results.get_forecast(steps=5)
forecast_index = pd.date_range(start=df.index[-1] + pd.DateOffset(years=1), periods=5, freq='Y')

# Plot results
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Count'], label='Observed')
plt.plot(forecast_index, forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(forecast_index,
                 forecast.conf_int()['lower Count'],
                 forecast.conf_int()['upper Count'],
                 color='pink', alpha=0.3)
plt.title('ARIMA Forecast of Journey Metaphor Usage')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#TIMES SERIES ANALYSIS (LAG1 REGRESSION) OF FAMILY METAPHOR (NATIONHOOD) IN NIGERIAN POLITICAL DISCOURSE

import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from scipy import stats

# Data: Year and Family metaphor usage
data = {
    "Year": np.arange(1960, 2025),
    "Family_References": [
        0, 5, 1, 2, 0, 0, 4, 1, 0, 5,
        29, 2, 0, 1, 1, 0, 1, 7, 3, 0,
        0, 1, 0, 4, 5, 8, 15, 8, 5, 9,
        8, 10, 0, 8, 0, 4, 3, 9, 5, 5,
        9, 0, 16, 9, 0, 7, 1, 5, 1, 1,
        28, 16, 7, 10, 27, 26, 3, 6, 5, 3,
        5, 5, 3, 12, 11
    ]
}

# Create DataFrame and prepare for analysis
df = pd.DataFrame(data)
years = df['Year']  # Save Year column before indexing
df = df.set_index("Year")

# PROPER TIME-TREND ANALYSIS ADDITION)
# Create time variable from index
df['Time'] = years.values  # Or: df.reset_index(inplace=True) if keeping Year column

# Fit time-trend model
X = sm.add_constant(df['Time'])
model = sm.OLS(df['Family_References'], X).fit()

# Get statistics
B_time = model.params['Time']
SE_time = model.bse['Time']
t_time = model.tvalues['Time']
p_time = model.pvalues['Time']
rsq_time = model.rsquared

# Residual diagnostics
residuals_time = model.resid
shapiro_test = stats.shapiro(residuals_time)

# Autocorrelation check for residuals
fig, ax = plt.subplots(2, 1)
plot_acf(residuals_time, ax=ax[0])
plot_pacf(residuals_time, ax=ax[1], method='ywm')
plt.show()

# Report PROPER TIME-TREND RESULTS
print("\nTime-Trend Analysis Results:")
print(f"A regression model testing annual changes in family metaphors was {'significant' if p_time < 0.05 else 'non-significant'}")
print(f"Year coefficient: B = {B_time:.3f}, SE = {SE_time:.3f}, t = {t_time:.2f}, p = {p_time:.3f}")
print(f"Model: Y = {model.params['const']:.2f} + {B_time:.3f}*Year")
print(f"Explained variance: {rsq_time*100:.1f}%")
print(f"Residual normality: W = {shapiro_test.statistic:.2f}, p = {shapiro_test.pvalue:.3f}")
print("ACF/PACF of residuals show no autocorrelation" if
      all(abs(plot_acf(residuals_time, alpha=0.05).axes[0].lines[1].get_ydata()) < 1.96/np.sqrt(len(residuals_time)))
      else "Residual autocorrelation present")
print(f"Conclusion: {'Significant' if p_time < 0.05 else 'No'} evidence of temporal change in FAMILY metaphor usage (α=0.05)")
```
