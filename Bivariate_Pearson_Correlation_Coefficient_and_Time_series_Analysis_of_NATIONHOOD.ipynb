```python
#TIME and ATTENTION TO NATIONHOOD in Nigerian Political discourse (Chapter Four)

import numpy as np
from scipy.stats import pearsonr

# Data: Year and Standardised number of nationhood references
years = np.array([1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015, 2020])
nationhood_references = np.array([0.75, 1.38, 4.09, 4.31, 4.49, 6.39, 4.32, 5.47, 3.90, 2.09, 7.23, 3.83, 3.64])

# Calculate Pearson correlation
correlation_coefficient, p_value = pearsonr(years, nationhood_references)

print(f"Pearson correlation coefficient: {correlation_coefficient}")
print(f"P-value: {p_value}")
```
